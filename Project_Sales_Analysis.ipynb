### Distribution of Data using DistPlot

### Ship Mode Wise Analysis

### Segment Wise Analysis

##### Which customer segment has highest amount of sales and find out the most profitable segment.

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Profit']].sum()
df_segment.Sales = df_segment.Sales.apply(lambda x: round(x,2))
df_segment
```

```python
figure, axis = plt.subplots(1, 2, figsize=(12, 6))
# Plot Bar Plot for Best Selling Segment
subseg1 = sns.barplot(data = df_segment, x = df_segment.index, y = df_segment.Sales,palette='plasma', ax=axis[0])
subseg1.set_title("Best Selling Segment",fontsize=15)
subseg1.set_xticklabels(subseg1.get_xticklabels(),rotation = "vertical", size = 10)
# Plot Bar Plot for Most Profitable Segment
subseg2 = sns.barplot(data = df_segment, x = df_segment.index, y = df_segment.Profit,palette='plasma', ax=axis[1])
subseg2.set_title("Most Profitable Segment",fontsize=15)
subseg2.set_xticklabels(subseg2.get_xticklabels(),rotation = "vertical", size = 10)
# Set spacing between subplots
figure.tight_layout()
plt.show()
```

##### 'Consumer' customer segment is the best selling with maximum profit.

##### Which was the busiest Month-Year of the store till now?

```python
df['Year_month']=df['Order Date'].dt.strftime('%y-%m')
df.head(15)
```

```python
df_date=df['Year_month'].value_counts().to_frame()         
df_date.reset_index(inplace=True)
df_date=df_date.rename(columns={'Year_month':'Count','index':'Date'})
df_time=df_date.head(20)
df_time=df_time.sort_values('Date')
df_time
```

##### September,2017 was the Busiest Month-Year of the store.
